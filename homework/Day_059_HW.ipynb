### 用 digits (3個數字即可 , dataset.load_digits(n_class=3)) 資料嘗試跑 PCA


```python
from sklearn import datasets
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
import matplotlib.pyplot as plt
import numpy as np

digits = datasets.load_digits(n_class=3)
X = digits.data
y = digits.target
```

載入 相關套件 並 執行 PCA 實驗 ...

```python
pca = decomposition.PCA(n_components=3)
```

```python
pca.fit(X)
X_new = pca.transform(X)
```

```python
fig = plt.figure(1, figsize=(4, 3))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
plt.cla()


for name, label in [('0', 0), ('1', 1), ('2', 2)]:
    ax.text3D(X_new[y == label, 0].mean(),
              X_new[y == label, 1].mean() + 1.5,
              X_new[y == label, 2].mean(), name,
              horizontalalignment='center',
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))
# Reorder the labels to have colors matching the cluster results
color = np.choose(y, [1, 2, 0]).astype(np.float)
ax.scatter(X_new[:, 0], X_new[:, 1], X_new[:, 2], c=color, cmap=plt.cm.rainbow,
           edgecolor='k')

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])

plt.show()
```

```python
## 利用PCA轉換成二維資料呈現 -> 顏色分布明顯有分群效果
pca_two = decomposition.PCA(n_components=2)
```

```python
pca_two.fit(X)
X_new = pca_two.transform(X)
```

```python
fig, ax = plt.subplots(1, 1)
fig.set_size_inches(18, 7)
 
for name, label in [('0', 0), ('1', 1), ('2', 2)]:
    ax.text(X_new[y == label, 0].mean(),
              X_new[y == label, 1].mean() + 1.5, name, c = 'red', size = '20',
              horizontalalignment='center',
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))
# Reorder the labels to have colors matching the cluster results
color = np.choose(y, [1, 0, 2]).astype(np.float)
ax.scatter(X_new[:, 0], X_new[:, 1], c=color, marker='.')


plt.show()
```
